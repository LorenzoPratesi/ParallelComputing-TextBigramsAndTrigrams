### Load Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
seq_2grams = pd.read_csv('/Users/lorenzopratesi/Documents/development/IntelliJ/PC-TextBigramsAndTrigrams/src/main/java/lorenzopratesi/benchmarks/results/bigrams/sequential.csv')
seq_2grams["Score"] = seq_2grams["Score"].str.replace(",",".").astype(float)
```

```python
parallel_2grams = pd.read_csv('/Users/lorenzopratesi/Documents/development/IntelliJ/PC-TextBigramsAndTrigrams/src/main/java/lorenzopratesi/benchmarks/results/bigrams/parallel.csv')
parallel_2grams["Score"] = parallel_2grams["Score"].str.replace(",",".").astype(float)
parallel_2grams
```

```python
parallel_2grams_2th = parallel_2grams.loc[parallel_2grams['Param: numOfThreads'] == 2]
parallel_2grams_4th = parallel_2grams.loc[parallel_2grams['Param: numOfThreads'] == 4]
parallel_2grams_6th = parallel_2grams.loc[parallel_2grams['Param: numOfThreads'] == 6]
parallel_2grams_8th = parallel_2grams.loc[parallel_2grams['Param: numOfThreads'] == 8]
parallel_2grams_10th = parallel_2grams.loc[parallel_2grams['Param: numOfThreads'] == 10]
parallel_2grams_12th = parallel_2grams.loc[parallel_2grams['Param: numOfThreads'] == 12]
```

```python
plt.figure(figsize=(13, 10))
plt.title('Parallel computing Bigrams over different file sizes', fontdict={'fontweight' : 'bold', 'fontsize': 14})

x_axis = np.log10([50,500,10000,50000,100000,150000]);

plt.plot(x_axis, seq_2grams.Score, 'k.-', label='Sequential')


plt.plot(x_axis, parallel_2grams_2th.Score, 'b.-', label='2 Thread')
plt.plot(x_axis, parallel_2grams_4th.Score, 'g.-', label='4 Thread')
plt.plot(x_axis, parallel_2grams_6th.Score, 'r.-', label='6 Thread')
plt.plot(x_axis, parallel_2grams_8th.Score, 'c.-', label='8 Thread')
plt.plot(x_axis, parallel_2grams_10th.Score, 'm.-', label='10 Thread')
plt.plot(x_axis, parallel_2grams_12th.Score, 'y.-', label='12 Thread')

plt.xticks(x_axis, ['0.05', '0.5', '10', '50', '100', '150'])


plt.xlabel('File sizes (Mb)')
plt.ylabel('Execution time (ms/op)')

plt.legend()
plt.show()
```

```python
speedup_2grams_2th = np.divide(seq_2grams.Score.tolist(), parallel_2grams_2th.Score.tolist())
speedup_2grams_4th = np.divide(seq_2grams.Score.tolist(), parallel_2grams_4th.Score.tolist())
speedup_2grams_6th = np.divide(seq_2grams.Score.tolist(), parallel_2grams_6th.Score.tolist())
speedup_2grams_8th = np.divide(seq_2grams.Score.tolist(), parallel_2grams_8th.Score.tolist())
speedup_2grams_10th = np.divide(seq_2grams.Score.tolist(), parallel_2grams_10th.Score.tolist())
speedup_2grams_12th = np.divide(seq_2grams.Score.tolist(), parallel_2grams_12th.Score.tolist())
```

```python
plt.figure(figsize=(13, 10))
plt.title('Speed Up Bigrams over different file sizes', fontdict={'fontweight' : 'bold', 'fontsize': 14})

x_axis = np.log10([50,500,10000,50000,100000,150000]);

plt.plot(x_axis, speedup_2grams_2th, 'b.-', label='2 Thread')
plt.plot(x_axis, speedup_2grams_4th, 'g.-', label='4 Thread')
plt.plot(x_axis, speedup_2grams_6th, 'r.-', label='6 Thread')
plt.plot(x_axis, speedup_2grams_8th, 'c.-', label='8 Thread')
plt.plot(x_axis, speedup_2grams_10th, 'm.-', label='10 Thread')
plt.plot(x_axis, speedup_2grams_12th, 'y.-', label='12 Thread')

plt.xticks(x_axis, ['0.05', '0.5', '10', '50', '100', '150'])


plt.xlabel('File sizes (Mb)')
plt.ylabel('Speed Up')

plt.legend()
plt.show()
```
