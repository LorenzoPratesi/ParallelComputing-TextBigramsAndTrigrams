```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
seq_2grams = pd.read_csv('/Users/lorenzopratesi/Documents/development/IntelliJ/PC-TextBigramsAndTrigrams/src/main/java/lorenzopratesi/benchmarks/results/bigrams/sequential.csv')
seq_2grams["Score"] = seq_2grams["Score"].str.replace(",",".").astype(float)
```

```python
seq_2grams_50KB = seq_2grams.loc[seq_2grams['Param: files'] == '1-text50KB']
seq_2grams_500KB = seq_2grams.loc[seq_2grams['Param: files'] == '2-text500KB']
seq_2grams_10MB = seq_2grams.loc[seq_2grams['Param: files'] == '3-text10MB']
seq_2grams_50MB = seq_2grams.loc[seq_2grams['Param: files'] == '4-text50MB']
seq_2grams_100MB = seq_2grams.loc[seq_2grams['Param: files'] == '5-text100MB']
seq_2grams_150MB = seq_2grams.loc[seq_2grams['Param: files'] == '6-text150MB']
```

```python
parallel_2grams = pd.read_csv('/Users/lorenzopratesi/Documents/development/IntelliJ/PC-TextBigramsAndTrigrams/src/main/java/lorenzopratesi/benchmarks/results/bigrams/parallel.csv')
parallel_2grams["Score"] = parallel_2grams["Score"].str.replace(",",".").astype(float)
```

```python
parallel_2grams_50KB = parallel_2grams.loc[parallel_2grams['Param: files'] == '1-text50KB']
parallel_2grams_500KB = parallel_2grams.loc[parallel_2grams['Param: files'] == '2-text500KB']
parallel_2grams_10MB = parallel_2grams.loc[parallel_2grams['Param: files'] == '3-text10MB']
parallel_2grams_50MB = parallel_2grams.loc[parallel_2grams['Param: files'] == '4-text50MB']
parallel_2grams_100MB = parallel_2grams.loc[parallel_2grams['Param: files'] == '5-text100MB']
parallel_2grams_150MB = parallel_2grams.loc[parallel_2grams['Param: files'] == '6-text150MB']
```

```python
plt.figure(figsize=(13, 10))
plt.title('Parallel computing Bigrams over different file sizes', fontdict={'fontweight' : 'bold', 'fontsize': 14})

x_axis = [2, 4, 6, 8, 10, 12];

#plt.plot(x_axis, seq_2grams_50KB, 'k.-', label='Sequential')

plt.plot(x_axis, parallel_2grams_50KB.Score, 'b.-', label='50KB')
plt.plot(x_axis, parallel_2grams_500KB.Score, 'g.-', label='500KB')
plt.plot(x_axis, parallel_2grams_10MB.Score, 'r.-', label='10MB')
plt.plot(x_axis, parallel_2grams_50MB.Score, 'm.-', label='50MB')
plt.plot(x_axis, parallel_2grams_100MB.Score, 'c.-', label='100MB')
plt.plot(x_axis, parallel_2grams_150MB.Score, 'y.-', label='150MB')

plt.xlabel('Number of Threads')
plt.ylabel('Execution time (ms/op)')

plt.legend()
plt.show()
```

```python
speedup_2grams_50KB = np.divide(seq_2grams_50KB.Score.tolist(), parallel_2grams_50KB.Score.tolist())
speedup_2grams_500KB = np.divide(seq_2grams_500KB.Score.tolist(), parallel_2grams_500KB.Score.tolist())
speedup_2grams_10MB = np.divide(seq_2grams_10MB.Score.tolist(), parallel_2grams_10MB.Score.tolist())
speedup_2grams_50MB = np.divide(seq_2grams_50MB.Score.tolist(), parallel_2grams_50MB.Score.tolist())
speedup_2grams_100MB = np.divide(seq_2grams_100MB.Score.tolist(), parallel_2grams_100MB.Score.tolist())
speedup_2grams_150MB = np.divide(seq_2grams_150MB.Score.tolist(), parallel_2grams_150MB.Score.tolist())
```

```python
plt.figure(figsize=(13, 10))
plt.title('Speed Up Bigrams over a different number of treads', fontdict={'fontweight' : 'bold', 'fontsize': 14})

x_axis = [2, 4, 6, 8, 10, 12];

plt.plot(x_axis, speedup_2grams_50KB, 'b.-', label='50KB')
#plt.plot(x_axis, speedup_2grams_500KB, 'g.-', label='500KB')
plt.plot(x_axis, speedup_2grams_10MB, 'r.-', label='10MB')
#plt.plot(x_axis, speedup_2grams_50MB, 'c.-', label='50MB')
plt.plot(x_axis, speedup_2grams_100MB, 'm.-', label='100MB')
#plt.plot(x_axis, speedup_2grams_150MB, 'y.-', label='150MB')



plt.xlabel('Number of Threads')
plt.ylabel('Speed Up')

plt.legend()
plt.show()
```
